```python
# using the spotify data:

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
data = pd.read_csv('spotify_data.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data.groupby('year').size()
```

```python
sns.scatterplot(data = data,y = 'popularity', x = range(len(data)))
```

```python
# how does popularity vary by decade?

sns.histplot(data = data, x = 'popularity', bins = np.arange(min(data['popularity'])-5,max(data['popularity']) +5,10))
```
